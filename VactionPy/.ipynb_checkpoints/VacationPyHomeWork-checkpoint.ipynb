```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import time
import requests
import gmaps
import os

# Import API key
from api_keys import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Define file and path where data are stored to perform analysis 
csv_path = "output_data/cities.csv"
vacation_data_df = pd.read_csv(csv_path)

vacation_data_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude 
locations = vacation_data_df[["Latitude", "Longitude"]]
Humidity = vacation_data_df["Humidity"].astype(float)
```

```python
#Plot Heatmap

fig = gmaps.figure(layout={
        'width': '610px',
        'height': '600px',
        'padding': '3px',
        'border': '2px solid black'})

# Create heat layer,
heat_layer = gmaps.heatmap_layer(locations, weights=Humidity,dissipating=False,max_intensity=10,
                                 point_radius=1)
#Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
